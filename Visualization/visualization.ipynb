# Cartesian Product and Expansion
## 1D, 2D, and 3D

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-2, 2, 200)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot(x, np.zeros_like(x), 'k-', linewidth=1)
plt.scatter(x, np.zeros_like(x), c=x, cmap='rainbow', s=5)
plt.xlabel('x')
plt.title('1D Space')
plt.grid(True)


x, y = np.meshgrid(x, x)

plt.subplot(1, 2, 2)
plt.scatter(x, y, c=x, cmap='rainbow', s=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Space (Cartesian Product)')
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 12))
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
z = np.linspace(-2, 2, 50)
x, y, z = np.meshgrid(x, y, z)

ax = plt.subplot(1, 3, 3, projection='3d')
ax.scatter(x, y, z, c=x, cmap='rainbow', s=5, alpha=0.1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Space (Cartesian Product)')
ax.view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()
```

```python
#f(x, y) = x + y
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = X + Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Graph of f(x, y) = x + y')
plt.show()
```

## Boolean Space

```python
# Adjusting the previous code to visualize only B^1 to B^3.

# Define the vertices of a unit square and cube
vertices_2d = list(product([0, 1], repeat=2))
vertices_3d = list(product([0, 1], repeat=3))

# Create the figure
fig = plt.figure(figsize=(15, 5))

# Add subplot for B^1 (a line segment)
ax1 = fig.add_subplot(131)
ax1.set_title('1D Boolean Domain $\\mathbb{B}^1$')
ax1.plot([0, 1], [0, 0], 'bo-')  # Plotting a line for 1D
ax1.text(0, 0, '(0)', ha='center', va='bottom', fontsize=8)
ax1.text(1, 0, '(1)', ha='center', va='bottom', fontsize=8)
ax1.set_xlim(-0.5, 1.5)
ax1.set_ylim(-0.5, 0.5)
ax1.set_xticks([0, 1])
ax1.set_yticks([])
ax1.set_aspect('equal')

# Add subplot for B^2 (a square)
ax2 = fig.add_subplot(132)
ax2.set_title('2D Boolean Domain $\\mathbb{B}^2$')
for s, t in vertices_2d:
    ax2.text(s, t, f'({s}{t})', ha='center', va='center')
ax2.set_xlim(-0.5, 1.5)
ax2.set_ylim(-0.5, 1.5)
ax2.set_xticks([0, 1])
ax2.set_yticks([0, 1])
ax2.set_aspect('equal')
ax2.grid(True)

# Add subplot for B^3 (a cube)
ax3 = fig.add_subplot(133, projection='3d')
ax3.set_title('3D Boolean Domain $\\mathbb{B}^3$')
for (s, t, u) in vertices_3d:
    ax3.text(s, t, u, f'({s}{t}{u})', ha='center', va='center')
ax3.set_xlim(-0.5, 1.5)
ax3.set_ylim(-0.5, 1.5)
ax3.set_zlim(-0.5, 1.5)
ax3.set_xticks([0, 1])
ax3.set_yticks([0, 1])
ax3.set_zticks([0, 1])
ax3.view_init(30, 30)

# Show the plots
plt.tight_layout()
plt.show()

```
